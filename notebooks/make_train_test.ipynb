## Создаем train/test выборки на исходных данных для дальнейшей генерации BIO-файлов

1. Подгружаем файл с шаблонами
2. Разделяем Шаблоны на Train/Test
3. На основе Test создаем две тестовые непересекающиеся выборки
4. На основе Train делаем train/validation

Аналогично делаем с файлом по организациям

```python
from sklearn.model_selection import train_test_split
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd().parent
DATA_DIR = ROOT_DIR.joinpath('data')
```

### Разбиваем набор шаблонов на train/test

```python
df_patterns = pd.read_excel(DATA_DIR.joinpath("interim/patterns.xlsx"), sheet_name='Шаблоны строк')
```

```python
df_patterns.info()
```

```python
train, test = train_test_split(df_patterns, random_state=86, test_size=0.3, shuffle=True)
```

```python
train['Шаблоны'].unique()
```

```python
test['Шаблоны'].unique()
```

```python
train.to_excel(DATA_DIR.joinpath('processed/train_patterns.xlsx'), index=False)
```

```python
test1, test2 = train_test_split(test, random_state=86, test_size=0.5, shuffle=True)
```

```python
test1.to_excel(DATA_DIR.joinpath('processed/test1_patterns.xlsx'), index=False)
test2.to_excel(DATA_DIR.joinpath('processed/test2_patterns.xlsx'), index=False)
```
